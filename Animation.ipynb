```python
import geemap
import ee
import geemap.geemap as map
```

```python
ee.Authenticate()
ee.Initialize(project='rohitshinde4680')
```

```python
m=map.Map()
m
```

```python
India = ee.FeatureCollection('projects/rohitshinde4680/assets/nationalBoundary')
m.addLayer(India,{},'boundary')
m
```

```python
region = ee.Geometry.Polygon(
    [[[67.060547, 4.740675],
     [67.060547, 38.065392],
     [98.964844, 38.065392],
     [98.964844, 4.740675],
     [67.060547, 4.740675]]],
    None,
    False)
```

```python
ndvi = ee.ImageCollection("MODIS/061/MOD13A2").select('NDVI')
m
```

```python
ndvi_india = ndvi.map(lambda img: img.set ('doy',
ee.Date(img.get('system:time_start')).getRelative('day', 'year')));
```

```python
distinctDOY = ndvi_india.filterDate('2015-02-18', '2016-02-18');
```

```python
filter = ee.Filter.equals(leftField='doy',rightField='doy')
join = ee.Join.saveAll('doy_matches')
joincol = ee.ImageCollection(join.apply(distinctDOY,ndvi_india,filter))
```

```python
comp= joincol.map(lambda img: ee.ImageCollection.fromImages(img.get('doy_matches')).reduce(ee.Reducer.mean()).set('doy',ee.Number(img.get('doy'))))
```

```python
ndviVis = {
  'min': 2000,
  'max': 10000,
  'palette': [
    'ffffff', 'ce7e45', 'df923d', 'f1b555', 'fcd163', '99b718', '74a901',
    '66a000', '529400', '3e8601', '207401', '056201', '004c00', '023b01',
    '012e01', '011d01', '011301'
  ],
};
```

```python
rgbvis= comp.map(lambda img: img.visualize(bands = ['NDVI_mean'],**ndviVis).clip(India))
```

```python
gifParams = {
  'region': region,
  'dimensions': 600,
  'crs': 'EPSG:4326',
  'framesPerSecond': 10,
  'format':'gif'
}
```

```python
print(rgbvis.getVideoThumbURL(gifParams))
```
